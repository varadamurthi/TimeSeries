```python
import pandas_datareader as pdr
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_msft = pdr.get_data_yahoo('MSFT')

df_msft.head()
```

```python
df_msft.tail()
```

```python
df_msft.resample(rule='A').max()
```

```python
df_msft.loc[df_msft['High']==df_msft['High'].max()] 
```

```python
df_msft.loc[df_msft['High']==df_msft['High'].min()] 
```
